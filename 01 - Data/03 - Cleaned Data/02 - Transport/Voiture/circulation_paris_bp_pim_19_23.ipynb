Import des libs

```python
import numpy as np
import pandas as pd
import os
```

```python
filepath = os.path.join('..', '..', '..', '02 - Raw Data', '02 - Transport', 'Transport et pollution', 'circulation_paris_bp_pim_19_23_raw.csv')
df = pd.read_csv(filepath, sep=";")
df.head()
```

```python
df['VOLUME_PIM'] = df['VOLUME_PIM'].round(2)
df['VITESSE_PIM'] = df['VITESSE_PIM'].round(2)
df['VITESSE_BP'] = df['VITESSE_BP'].round(2)
df['VPBP_ANNUEL'] = df['VPBP_ANNUEL'].round(2)
df['KBP_ANNUEL'] = df['KBP_ANNUEL'].astype(int)
df.head()
```

```python
df.to_csv("circulation_paris_bp_pim_19_23_clean.csv")
```
