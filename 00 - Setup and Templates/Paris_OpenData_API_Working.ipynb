## Accessing the Paris OpenData API

```python
# importing libs

import numpy as np
import pandas as pd
import os
```

```python
# Defining the desired dataset
dataset_id = 'espaces_verts'

# Accessing and loading the dataset from opendata.paris.fr
df = pd.read_csv(f'https://opendata.paris.fr/api/explore/v2.1/catalog/datasets/{dataset_id}/exports/csv?timezone=Europe%2FParis&use_labels=false&epsg=4326', sep=';')
print("Loaded {} rows and {} columns".format(df.shape[0], df.shape[1]))
df.head()
```
